```python
import pandas as pd
from dotenv import load_dotenv
from os import getenv
load_dotenv()
```

```python
df = pd.read_csv(getenv("FG_MAPPED_TSV"), sep="\t")
```

```python
df.head()
```

```python
cmbname_df = pd.read_excel(getenv("FG_FINAL_CURATED_NAMES_XLSX"))
```

```python
cmbname_df.keys()
```

```python
cmbname_df.index = cmbname_df["cluster"]
```

```python
cmbname_df["stdname"] = cmbname_df.name
mask = cmbname_df.stdname.duplicated(keep="first")
cmbname_df.loc[mask, "stdname"] = cmbname_df.loc[mask, "stdname"] \
    + "." \
    + cmbname_df.loc[mask, :].groupby('stdname').cumcount().add(1).astype("str")
```

```python
mrgdf = cmbname_df.loc[df.tierNname, ["name", "type", "stdname"]]\
    .rename({"name": "diffexpname",
             "stdname": "curatedname"}, axis=1)
```

```python
df = pd.concat([df.reset_index(drop=True), mrgdf.reset_index(drop=True)], axis=1)
```

```python
pd.crosstab(df.tierNname, df.curatedname)
```

```python
print((df.curatedname == "").mean())
print(df.curatedname.isna().mean())
```
