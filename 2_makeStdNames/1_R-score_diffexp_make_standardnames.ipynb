```r
library(openxlsx)
library(tidyverse)
dotenv::load_dot_env()
```

# CD 

```r
xlsxfile = Sys.getenv("CD14_1vrest_xlsx")
sheets = getSheetNames(xlsxfile)
```

```r
cd14diffexp = sapply(sheets, function(s) read.xlsx(xlsxfile, sheet=s), simplify=FALSE, USE.NAMES = TRUE)
```

```r
cd14diffexp %>% class
```

```r
cd14diffexp = data.table::rbindlist(cd14diffexp, idcol="celltype")
```

```r
cd14diffexp %>% 
    mutate(rnkscr = -log1p(p_val_adj) * avg_logFC * (pct.1 / pct.2)) %>%
    group_by(cluster) %>%
#     arrange(cluster, desc(rnkscr))
    top_n(5, wt=rnkscr)
```

```r
cd14diffexp = separate(cd14diffexp, col=celltype, into=c("dataset", "type"), remove=FALSE)
cd14diffexp[grep(".*([Dd]oublet)[^\\.]*$", cd14diffexp$cluster), "type"] = "Doublets"

cdnames = cd14diffexp %>% 
    mutate(rnkscr = -log(p_val_adj+1e-310) * avg_logFC * (pct.1 / (pct.2 + 1e-300))) %>%
    group_by(cluster) %>%    
    top_n(2, wt=rnkscr) %>%
    arrange(cluster, desc(rnkscr)) %>%
    summarize(
        avg_rnkscr=mean(rnkscr),
        dataset=first(dataset),
        type=first(type),
        name=paste0(dataset, ".", type, paste0(".", gsub("\\.", "", gene), collapse="")))
```

```r
curatedtclls_df = read.xlsx(Sys.getenv("CD_curated_tcells_xlsx"))
curatedmloid_df = read.xlsx(Sys.getenv("CD_curated_myeloid_xlsx"))
```

```r
cdnames$name[match(curatedtclls_df$cluster[!is.na(curatedtclls_df$Combined.Name)], cdnames$cluster)] = 
    paste0("CD.", curatedtclls_df$Combined.Name[!is.na(curatedtclls_df$Combined.Name)])
cdnames$name[match(curatedmloid_df$cluster[!is.na(curatedmloid_df$Combined.Name)], cdnames$cluster)] = 
    paste0("CD.", curatedmloid_df$Combined.Name[!is.na(curatedmloid_df$Combined.Name)])
```

```r
cdnames$stdname = make.unique(cdnames$name)
```

```r
cdnames
```

```r
cdnames %>%
    write.table("20200625_CD_names.tsv", sep="\t", row.names = FALSE, quote=FALSE)
```

# FGID

```r
xlsxfile = Sys.getenv("FG13_1vrest_xlsx")
sheets = getSheetNames(xlsxfile)
```

```r
FG13diffexp = sapply(sheets, function(s) read.xlsx(xlsxfile, sheet=s), simplify=FALSE, USE.NAMES = TRUE)
```

```r
FG13diffexp = data.table::rbindlist(FG13diffexp, idcol="celltype", fill=TRUE)
```

```r
FG13diffexp$cluster[is.na(FG13diffexp$cluster)] = "Hs.Mstcl"
```

```r
FG13diffexp = separate(FG13diffexp, col=celltype, into=c("dataset", "type"), remove=FALSE)
FG13diffexp[grep(".*([Dd]oublet)[^\\.]*$", FG13diffexp$cluster), "type"] = "Doublets"
FG13diffexp$dataset = "FG"

fgnames = FG13diffexp %>% 
    mutate(rnkscr = -log(p_val_adj+1e-300) * avg_logFC * (pct.1 / (pct.2 + 1e-300))) %>%
    group_by(cluster) %>%    
    top_n(2, wt=rnkscr) %>%
    arrange(cluster, desc(rnkscr)) %>%
    summarize(
        avg_rnkscr=mean(rnkscr),
        dataset=first(dataset),
        type=first(type),
        name=paste0(dataset, ".", type, paste0(".", gsub("\\.", "", gene), collapse="")))
```

```r
fgcuratedtclls_df = read.xlsx(Sys.getenv("FG_curated_tcells_xlsx"))
fgcuratedmloid_df = read.xlsx(Sys.getenv("FG_curated_myeloid_xlsx"))
```

```r
fgnames$name[match(fgcuratedtclls_df$cluster[!is.na(fgcuratedtclls_df$Combined.Name)], fgnames$cluster)] = 
    paste0("FG.", fgcuratedtclls_df$Combined.Name[!is.na(fgcuratedtclls_df$Combined.Name)])
fgnames$name[match(fgcuratedmloid_df$cluster[!is.na(fgcuratedmloid_df$Combined.Name)], fgnames$cluster)] = 
    paste0("FG.", fgcuratedmloid_df$Combined.Name[!is.na(fgcuratedmloid_df$Combined.Name)])
```

```r
fgnames$stdname = make.unique(fgnames$name)
```

```r
fgnames
```

```r
fgnames %>%
    write.table("20200625_FG_names.tsv", sep="\t", row.names = FALSE, quote=FALSE)
```
